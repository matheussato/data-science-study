```python
import pandas as pd
```

```python
dados = pd.read_excel('POP2022_Municipios.xls', header=1,skipfooter=34)
dados
```

```python
dados = dados.assign(populacao_sem_parenteses = dados['POPULAÇÃO'].replace('\(\d{1,2}\)', '',regex = True),
    populacao = lambda x : x.loc[:, 'populacao_sem_parenteses'].replace('\.', '', regex = True),)
```

```python
dados
```

```python
dados = dados.astype({'populacao':'int64'})
```

```python
dados = dados.groupby('UF')[['populacao']].sum().reset_index()
dados
```
